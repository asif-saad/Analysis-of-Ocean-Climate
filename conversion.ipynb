```python
## renaming operation

def renaming(data):
    data.rename(columns={'longitude': 'longitude_degrees_east',
                    'latitude':'latitude_degrees_north',
                    'pressure': 'Pressure_dbar',
                    'temperature': 'Temperature_Celcius',
                    'sigma_theta':'Sigma_theta_kg_m_pow_minus_3',
                    'depth':'depth_m',
                    'oxygen':'oxygen_mu_molkg_pow_minus_1',
                    'oxygen saturation': 'oxygen_saturation_percentage',
                    'nitrate': 'nitrate_mu_mol_kg_pow_minus_1',
                    'chl_a': 'chl_a_mg_m_pow_minus_3',
                    'b_bp700': 'b_bp700_m_pow_minus_1',
                    'POC': 'POC_milli_mol_m_pow_minus_3',
                    'pH_insitu': 'pH_insitu_total',
                    'pH_25C': 'pH_25C_total',
                    'TALK': 'TALK_micro_mol_kg_pow_minus_1',
                    'DIC': 'DIC_micro_mol_kg_pow_minus_1',
                    'pCO2': 'pCO2_micro_atm',
                    'b_bp532':'b_bp532_m_pow_minus_1',
                    'CDOM':'CDOM_ppb',
                    'salinity': 'salinity_pss',
                    'oxygen saturation_QF': 'oxygen_saturation_QF'},inplace=True)
    return data


```

```python
main=renaming(main)
```

```python
## checking the order of columns after reordering with another csv file dataset


colNorm= main.columns
colMod=reorder_data.columns
flag=True
for i in range(len(colMod)):
    if(colMod[i]!=colNorm[i]):
        flag=False
        print("something wrong here:")
        print("index: ",i," column: ",colMod[i]," and ",colNorm[i])
        
if(flag):
    print("everything in good order!!")
```
